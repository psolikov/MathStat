```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.5
c = 1 / (2*a + 2*np.exp(-a))

#Density
def f(x):
    if (x < -a):
        return c*np.exp(x)
    elif (x > a):
        return c*np.exp(-x)
    else:
        return c
    
def F(x):
    if x < -a:
        return c*np.exp(x)
    elif x < a:
        return c*(np.exp(-a) + a + x)
    else:
        return c*(2*np.exp(-a) + 2*a - np.exp(-x))
    
def F_inverse(x):
    inflection = c*np.exp(-a)
    if (x < inflection):
        return np.log(x / c)
    elif (x < inflection + c*2*a):
        return x/c - a - np.exp(-a)
    else:
        return -np.log(-x/c + 2*a + 2*np.exp(-a))
    
def F_decomp(x):
    inflection = c*np.exp(-a)
    ch = np.random.rand()
    if ch < inflection:
        return -a - np.random.exponential()
    elif ch < (inflection + c*2*a):
        return a*(2*x - 1)
    else:
        return a + np.random.exponential()
```

```python
### Inverse method with 50 dots
dots = 50
X = [F_inverse(x) for x in np.random.rand(dots)]
Xf = [i/100 for i in range(-550, 550)]
Yf = [F(x) for x in Xf]
XF = np.sort(X)
YF = [len(XF[XF < i/100]) / dots for i in range(-550, 550)]
plt.plot(Xf, Yf)
plt.plot(Xf, YF)
plt.show()
```
